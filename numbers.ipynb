```python
import pandas as pd
import requests
from splinter import Browser
from bs4 import BeautifulSoup as bs
```

```python
# Initialise df with homepage table
url = 'https://www.the-numbers.com/movie/budgets/all'
response = requests.get(url)
tables = pd.read_html(response.text)
budget_df = tables[0]
```

```python
budget_df
```

```python
# Iterate through all available urls to obtain full dataset
for i in range(1, 61):
    it_url = 'https://www.the-numbers.com/movie/budgets/all/' + str(i*100+1)
    response = requests.get(it_url)
    tables = pd.read_html(response.text)
    df2 = tables[0]
    print(it_url)

    budget_df = budget_df.append(df2, ignore_index=True) 
```

```python
budget_df
```

```python
clean_df = budget_df.iloc[:, 1:6]

clean_df['ProductionBudget'] = clean_df['ProductionBudget'].str.replace(',', '').\
    str.replace('$', '').astype(int)
clean_df['DomesticGross'] = clean_df['DomesticGross'].str.replace(',', '').\
    str.replace('$', '').astype(int)
clean_df['WorldwideGross'] = clean_df['WorldwideGross'].str.replace(',', '').\
    str.replace('$', '').astype(int)
```

```python
clean_df
```

```python
clean_df['profit_percent'] = ((clean_df['WorldwideGross']-clean_df['ProductionBudget'])/clean_df['ProductionBudget'])*100
clean_df
```

```python
clean_df['ReleaseYear'] = ""
for index, row in clean_df.iterrows():
    if "," in str(row['ReleaseDate']):
#         row['ReleaseYear'] = str(row['ReleaseDate']).rsplit(',', 1)[1]
        clean_df.loc[index, 'ReleaseYear'] = str(row['ReleaseDate']).rsplit(',', 1)[1]
        
    elif 'Unknown' in str(row['ReleaseDate']):
        row['ReleaseDate'] = str(row['ReleaseDate']).replace('Unknown', '')
    else:
        pass

clean_df['ReleaseYear'] = clean_df['ReleaseYear'].str.replace(' ', '').\
    str.replace(' ', '')
```

```python
clean_df.dtypes
```

```python
renamed_df = clean_df.rename(columns={'ProductionBudget':'production_budget', 'DomesticGross':'domestic_gross',
                                      'WorldwideGross':'worldwide_gross'})
```

```python
# saved scraped file as csv for future use
renamed_df.to_csv('renamed.csv', index=False)
```

```python
# read in imdb movies list from Kaggle
df = pd.read_csv('Resources/imdb_movies.csv', low_memory=False)
```

```python
movies_df = df[['imdb_title_id', 'original_title', 'year', 'genre', 'avg_vote']]
movies_df.head()
```

```python
import jellyfish
```

```python
# make copy of movies_df to transform
clean_df.head()
final_df = movies_df.copy()
```

```python
final_df.head()
```

```python
# created additional columns in final_df
final_df['production_budget'] = ""
final_df['domestic_gross'] = ""
final_df['worldwide_gross'] = ""
final_df['profit_percent'] = ""
```

```python
for index, row in final_df.iterrows():
    title = row['original_title']
    year = row['year']
    
    best_match = None
    highest_jw = 0
    for index_2, row_2 in renamed_df.iterrows():
        score = jellyfish.jaro_winkler(title.lower(), row_2['Movie'].lower())
        if (score > highest_jw and score > 0.9 and str(year)==str(row_2['ReleaseYear'])):
            highest_jw = score
            production_budget = row_2['production_budget']
            domestic_gross = row_2['domestic_gross']
            worldwide_gross = row_2['worldwide_gross']
            profit_percent = row_2['profit_percent']

    if (best_match):
        final_df.loc[index,'production_budget'] = production_budget
        final_df.loc[index,'domestic_gross'] = domestic_gross
        final_df.loc[index,'worldwide_gross'] = worldwide_gross
        final_df.loc[index,'profit_percent'] = profit_percent

    # Check matches from jellyfish
    print(index, title, best_match, highest_jw)
```

```python
# save complete table to csv for future reference
final_df.to_csv('movie_numbers.csv', index=False)
```
