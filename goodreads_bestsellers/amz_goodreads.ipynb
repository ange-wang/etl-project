```python
import pandas as pd
import jellyfish
```

```python
amz_df = pd.read_csv("amazon_books.csv")
best_sellers = amz_df.copy()
best_sellers = best_sellers.drop_duplicates('Name').rename({'Name': 'title'}, axis =1).sort_values('title', ascending = True)

#removing parenthesis
best_sellers['title'] = best_sellers['title'].apply(lambda x: x.split("(")[0])
#removing white space
best_sellers['title'] = best_sellers['title'].apply(lambda x: x.strip())

#removing more duplicates
best_sellers['title'] = best_sellers['title'].apply(lambda x: x.lower())
best_sellers = best_sellers.drop_duplicates('title') 
best_sellers['title'] = best_sellers['title'].apply(lambda x: x.title())

#sorting
best_sellers = best_sellers.sort_values('title', ascending = True)
best_sellers.head()
```

```python
goodreads = pd.read_csv("goodreads_books.csv")
goodread_books = goodreads.copy()
goodread_books = goodread_books[['title', 'authors', 'isbn', 'average_rating', 'ratings_count', 'text_reviews_count']].drop_duplicates('title')
goodread_books['best_seller'] = ""

#removing parenthesis
goodread_books['title'] = goodread_books['title'].apply(lambda x: x.split("(")[0])

#removing white space
goodread_books['title'] = goodread_books['title'].apply(lambda x: x.strip())

#more removal
goodread_books['title'] = goodread_books['title'].apply(lambda x: x.lower())
goodread_books = goodread_books.drop_duplicates('title')
goodread_books['title'] = goodread_books['title'].apply(lambda x: x.title())

#sorting 
goodread_books = goodread_books.sort_values('title', ascending = True).reset_index(drop = True)
goodread_books.head(30)
```

```python
#loop through both data frames 
#if titles are almost identical then fill-in best sellers column w/ True
for index_gr, row_gr in goodread_books.iterrows():
    
    #good_reads title name
    gr_title = row_gr['title']
    
    for index_bs, row_bs in best_sellers.iterrows():
        
        #best seller title name
        bs_title = row_bs['title']
        
        #jelly compare
        jelly_compare = jellyfish.jaro_winkler(gr_title, bs_title)
        
        if jelly_compare > 0.74:
            goodread_books['best_seller'].iloc[index_gr] = True
        
        #else:
        #    goodread_books['best_seller'].iloc[index_gr] = False
            
            
goodread_books.to_csv("Outputs/goodreads_and_bestsellers.csv")        
```
