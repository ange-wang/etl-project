```python
from bs4 import BeautifulSoup
from splinter import Browser
from pprint import pprint
import pymongo
import pandas as pd
import requests
```

```python
a_to_c = pd.read_html("https://en.wikipedia.org/wiki/List_of_fiction_works_made_into_feature_films_(0%E2%80%939,_A%E2%80%93C)")
a_to_c_movies = [a_to_c[1], a_to_c[2], a_to_c[3]]
a_to_c_movies_df = pd.concat(a_to_c_movies)
a_to_c_movies_df
```

```python
d_to_j = pd.read_html("https://en.wikipedia.org/wiki/List_of_fiction_works_made_into_feature_films_(D%E2%80%93J)")
d_to_j_movies = [d_to_j[1], d_to_j[2], d_to_j[3], d_to_j[4], d_to_j[5], d_to_j[6]]
d_to_j_movies_df = pd.concat(d_to_j_movies)
d_to_j_movies_df
```

```python
k_to_r = pd.read_html("https://en.wikipedia.org/wiki/List_of_fiction_works_made_into_feature_films_(K%E2%80%93R)")
k_to_r_movies = [k_to_r[1], k_to_r[2], k_to_r[3], k_to_r[4], k_to_r[5], k_to_r[6], k_to_r[7], k_to_r[8]]
k_to_r_movies_df = pd.concat(k_to_r_movies)
k_to_r_movies_df
```

```python
s_to_z = pd.read_html("https://en.wikipedia.org/wiki/List_of_fiction_works_made_into_feature_films_(S%E2%80%93Z)")
s_to_z_movies = [s_to_z[1], s_to_z[2], s_to_z[3], s_to_z[4], s_to_z[5], s_to_z[6]]
s_to_z_movies_df =  pd.concat(s_to_z_movies)
s_to_z_movies_df
```

```python
a_to_z_movies = [a_to_c_movies_df, d_to_j_movies_df, k_to_r_movies_df, s_to_z_movies_df]
a_to_z_movies_df = pd.concat(a_to_z_movies)
a_to_z_movies_df
```

```python
#set delimiter to be used to split movies adapted from books multiple times
a_to_z_movies_df["Film adaptation(s)"] = a_to_z_movies_df["Film adaptation(s)"].str.replace(f")", f"),")
```

```python
#rename columns
a_to_z_movies_df = a_to_z_movies_df.rename(columns={"Fiction work(s)": "Books", "Film adaptation(s)":"Movies"})
a_to_z_movies_df
```

```python
#explode books then save to excel for excel data cleaning
a_to_z_movies_df = a_to_z_movies_df.assign(Books=a_to_z_movies_df.Books.str.split("*")).explode("Books")
a_to_z_movies_df.to_csv("clean_a_to_z.csv", encoding="utf-8", index=False)
```

```python
#read drop blank rows and save overwrite the cleaned file
clean_a_to_z = pd.read_csv('clean_a_to_z.csv') 
clean_a_to_z = clean_a_to_z[clean_a_to_z['Books'].notna()]
clean_a_to_z.to_csv("clean_a_to_z.csv", encoding="utf-8", index=False)
clean_a_to_z
```

```python
#Explode Movies
test_db = clean_a_to_z.assign(Movies=clean_a_to_z.Movies.str.split(",")).explode("Movies")
test_db
```

```python
#Clean data to select movie rows that doesn't contain blanks: then clean all rows with excel
#Then explode the year from movies using excel :P cause it's easier to do it there
test_db2 = test_db[test_db['Movies']!=""]
test_db2.to_csv("test3.csv", encoding="utf-8", index=False)
test_db2
```
